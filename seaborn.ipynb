```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(5)
```

```python
ax = sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm')
```

```python
fg = sns.relplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', kind='scatter')
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', kind='scatter', col='sex', hue='species')
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', kind='scatter', hue='body_mass_g')
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', kind='scatter', hue='bill_depth_mm')
```

```python
#  
flights = sns.load_dataset('flights')
flights.head(7)
```

```python
#      
sns.relplot(data=flights, x='year', y='passengers', kind='line')
```

```python
# 
sns.displot(data=penguins, x='body_mass_g', kind='hist')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', hue='species')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', hue='species', col='sex', row='island')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='kde', hue='species', fill=True)
```

```python
sns.displot(data=penguins, x='body_mass_g', y='bill_length_mm', cbar=True)
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species')
```

```python
sns.pairplot(data=penguins, hue='species')
```

```python
sns.catplot(data=penguins, x='species', y='flipper_length_mm')
```

```python
sns.catplot(data=penguins, x='species', y='flipper_length_mm', kind='box')
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm')
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='island')
#     =   
```
